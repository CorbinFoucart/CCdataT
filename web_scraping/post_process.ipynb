```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
collected_timeseries = []
months = [2, 3, 4, 5, 6]
for month in months:
    timeseries_file = 'justfly_LAX_HKG_prices_{}_2019.csv'.format(month)
    collected_timeseries.append(pd.Series.from_csv(timeseries_file))
```

```python
month_labels = ['feb', 'mar', 'apr', 'may', 'jun']
```

```python
collected_timeseries[2]
```

```python
# find best price 
good = []
for series in collected_timeseries:
    good.append(series.nsmallest(n=3))
    
```

```python
pd.concat(good)
```

```python
import seaborn as sns
sns.set()
```

```python
fig, ax = plt.subplots(figsize=(15, 6))

for idx, timeseries in enumerate(collected_timeseries):
    ax.plot(timeseries, '-o', label=month_labels[idx])

# rotate and align tick labels so they look better
fig.autofmt_xdate()
ax.set_xlabel('departure date')
ax.set_ylabel('price (USD)')
ax.set_title('week vacation: roundtrip flight prices: LAX -> HKG, JustFly')
plt.legend(bbox_to_anchor=(1,1))
plt.savefig('prices.png', bbox_inches='tight')
plt.show()
```
